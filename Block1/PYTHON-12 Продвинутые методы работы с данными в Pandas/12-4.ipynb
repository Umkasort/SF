```python
import pandas as pd
melb_data=pd.read_csv('data/melb_data_fe.csv')
melb_df=melb_data.copy()
melb_df.pivot_table(values='Price',index='Rooms',columns='Type',fill_value=0).round()
melb_df.pivot_table(values='Price',columns='Weekend',index='Regionname',aggfunc='count')
melb_df.pivot_table(values='Price',index=['SellerG'],columns='Type',aggfunc=['median'],fill_value=0)
```
