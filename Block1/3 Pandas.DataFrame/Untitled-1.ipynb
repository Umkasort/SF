```python
# Импортируем библиотеку для выполнения HTTP-запросов в интернет
import requests
# Читаем текстовый файл по url-ссылке
data = requests.get("https://raw.githubusercontent.com/SkillfactoryDS/Datasets/master/war_peace_processed.txt").text
# Предобрабатываем текстовый файл
data = data.split('\n')
data.remove('')
data = data + ['[new chapter]']
# Превращаем список в множество, удаляя дублирующиеся слова
word_set = set(data)
# Удаляем из множества слово, символизирующее раздел между главами
word_set.discard('[new chapter]')
# Инициализируем пустой словарь
word_counts = {}
# Инициализируем количество глав
count_chapter = 0
# Создаем цикл по всем словам из списка слов
for word in data:
    # Проверяем, что текущее слово - обозначение новой главы
    if word == '[new chapter]':
        # Если условие выполняется, то увеличиваем количество глав на 1
        count_chapter += 1
        # Переходим на новую итерацию цикла
        continue
    # Проверяем, что текущего слова еще нет в словаре слов
    if word not in word_counts:
        # Если условие выполняется, инициализируем новый ключ 1
        word_counts[word] = 1
    else:
        # В противном случае, увеличиваем количество слов на 1
        word_counts[word] += 1
# Создаем цикл по ключам и их порядковым номерам полученного словаря
for i, key in enumerate(word_counts):
    # Выводим только первые 10 слов
    if i == 10:
        break
# Инициализируем общий список, в котором будем хранить списки слов в каждой главе
chapter_data = []
# Инициализируем список слов, в котором будет хранить слова одной главы
chapter_words = []
# Создаем цикл по всем словам из списка
for word in data:
    # Проверяем, что текущее слово - обозначение новой главы
    if word == '[new chapter]':
        # Если условие выполняется, добавляем список со словами из главы в общий список
        chapter_data.append(chapter_words)
        # Обновляем (перезаписываем) список со словами из текущей главы
        chapter_words = []
    else:
        # В противном случае, добавляем текущее слово в список со словами из главы
        chapter_words.append(word)
# Инициализируем список, в котором будем хранить словари
chapter_words_count = []
# Создаем цикл по элементам внешнего списка со словами
for chapter_words in chapter_data:
    # Инициализируем пустой словарь, куда будем добавлять результаты
    temp = {}
    # Создаем цикл по элементам внутреннего списка
    for word in chapter_words:
        # Проверяем, что текущего слова еще нет в словаре
        if word not in temp:
            # Если условие выполняется, добавляем ключ в словарь
            temp[word] = 1
        else:
            # В противном случае, увеличиваем количество влождений слова в главу
            temp[word] += 1
# Добавляем получившийся словарь в список
    chapter_words_count.append(temp)
print (word_set)
print( count_chapter)
print(word_counts)
print(chapter_data [:2][:2])
print(chapter_words_count[:2][:1])
```

```python
input_string = input('Введите число словами: ')
# ваш код здесь
# Объявляем функцию
def transform_string_to_integer(text):
# Обрабатываем исключение из алгоритма (максимальное число)  
    if text == 'один миллион': return 1000000
# Задаем словарь прописных цифр   
    number_word = dict()
# Задаем словарь разрядов (пригодится, если использовать алгоритм при числах выше миллиона)
    razryad = {1000:"тыс"}
# Задаем прописные цифры по разряду единиц, объединив их с 11-19 (условно считаем их числами первого разряда) и сопоставляем значения
    number_word[1] = {"одиннадцать": 11, "двенадцать": 12, "тринадцать": 13, "четырнадцать": 14, "пятнадцать": 15, "шестнадцать": 16, 
                      "семнадцать": 17, "восемнадцать": 18, "девятнадцать": 19,"десять": 10, "ноль": 0, "один": 1, "два": 2, "три": 3, 
                      "четыре": 4, "пять": 5, "шесть": 6, "семь": 7, "восемь": 8, "девять": 9, 'одна': 1, 'две':2}
# То же самое для десятков (20-90)
    number_word[10] = {"двадцать": 20, "тридцать": 30, "сорок": 40, "пятьдесят": 50, "шестьдесят": 60, "семьдесят": 70, "восемьдесят": 80, 
                       "девяносто": 90}
# То же самое для сотен (100-900)
    number_word[100] = {"сто": 100, "двести": 200, "триста": 300, "четыреста": 400, "пятьсот": 500, "шестьсот": 600, "семьсот": 700, 
                        "восемьсот": 800, "девятьсот": 900}
# Создаем выходное число 
    res = 0
# Разбиваем исходный текст в слова
    text_lst = text.split()
# Счетчик разрядов в числе
    count = 1
# Счетчик разрядов в словаре
    count_razr=1
# Цикл по списку слов в обратном порядке (от единиц по возрастанию)
    for word in text_lst[::-1]:
# Ищем слово тысяча, если находим - продолжаем цикл
        if word.startswith(razryad[1000]):
            count = 1000
            continue
# Флаг поиска цифры в словаре
        flag_find = False
# Цикл по словарю по разрядам (1, 10, 100)
        for razr in number_word:
# Цикл по словарю внутри разряда (непосредственно сами цифры)
            for str_int in number_word[count_razr]:
# Если цифра найдена
                if word == str_int:
# Получаем цифровое ее представление
                    number = number_word[count_razr][str_int]
# Если разряды в тысячах, 
                    if count > 100:
# то плюсуем к выходному значению цифру умноженную на 1000
                        res += number * 1000
# иначе если от 1 до 999 
                    else:
# то просто плюсуем значение к выходному значению
                        res += number
# устанавливаем флаг поиска в истину                   
                    flag_find = True
# увеличиваем разрядность числа в 10 раз
                    count *= 10
# увеличиваем разрядность словаря в 10 раз
                    count_razr *= 10
# и прерываем цикл
                    break
# если разрядность словаря 1000, возвращаем к 1 
            if count_razr == 1000:count_razr =1
# если слово во вложенном цикле не найдено,
            if not flag_find:
# все равно увеличиваем разрядность числа в (0 в середине числа) 
                count *= 10
# и разрядность поиска счетчика разряда
                count_razr *= 10
# если слово найдено, прерываем и этот цикл
            if flag_find: break
# если слово было найдено, продолжаем цикл
            if flag_find: continue
# если счетчик разрядов выще миллиона, останавливаем цикл
        if count > 1000000: break
# возвращаем значение
    return res
print(transform_string_to_integer(input_string))
```
